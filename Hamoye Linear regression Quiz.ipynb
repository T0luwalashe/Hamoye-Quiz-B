### Importing the relevant libaries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
from sklearn.linear_model import LinearRegression
import seaborn as sns
sns.set()
```

```python
#oaiding the data set
data = pd.read_csv('energydata_complete_ (Appliances energy prediction Data Set).csv')
```

```python
## Lets create a copy of data which we would manipulate
df = data.copy()
```

```python
#Declaring variables of interest and dropping irrelevant data set 
df = df.drop(['date'], axis = 1)
df_cleaned = df.drop(['lights'], axis =1)
```

```python
##for question 12 a model between T2 and T2
# lets declare the inputs & targetS and fit the modell
targets = df_cleaned['T2']
inputs = df_cleaned['T6']
```

```python
x = sm.add_constant(inputs)
results = sm.OLS(targets,x).fit()
results.summary()
```

```python
# (Question13) we are trying to fit a model betwwen the temperature in the living room and outside 
#so lets define our targets and inputs

targets = df_cleaned['Appliances']
inputs = df_cleaned.drop(['Appliances'], axis =1)
```

```python
#lets standardize the variables
# using MinMaxScaler 
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(inputs)
inputs_scaled = scaler.transform(inputs)
```

```python
# lets use the ols method to fit the data
x = sm.add_constant(inputs_scaled)
results = sm.OLS(targets,x).fit()
results.summary()
```

```python
# For the mean absolute error, we have to fit the model and declare our test, train split
from sklearn.model_selection import train_test_split
x_train, x_test,y_train, y_test = train_test_split(inputs_scaled , targets , test_size = 0.3 ,random_state = 42)
```

```python
## Lets create the regression
reg = LinearRegression()
reg.fit(x_train, y_train)
```

```python
y_hat = reg.predict(x_train)
```

```python
#Mean Absolute Error
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_train, y_hat)
round(mae, 2) 
```

```python
## The root mean squared error
from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(y_hat, y_train))
round(rmse, 2)
```

```python
#the r2
from sklearn.metrics import r2_score
r2_score = r2_score(y_hat, y_train)
r2_score
```

```python
#To get the weights of each variables 
Linear = pd.DataFrame(inputs.columns.values, columns=['Features'])
Linear['Linear'] = reg.coef_
Linear
```

```python
## Ridge regresion on the test set
from sklearn.linear_model import Ridge
ridge_reg = Ridge(alpha = 0.4)
ridge_reg.fit(x_test,y_test)
```

```python
#Lasso regression on the data set

from sklearn.linear_model import Lasso
lasso_reg = Lasso(alpha = 0.001)
lasso_reg.fit(x_train,y_train)
```

```python
#getting the weights form the lasso regression
Lasso = pd.DataFrame(inputs.columns.values, columns=['Features'])
Lasso['Lasso'] = lasso_reg.coef_
Lasso
```
